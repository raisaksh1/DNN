```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/adm_data.csv')
```

```python
print(df.shape)
df.head()
```

```python
sns.scatterplot(data = df, x = 'TOEFL Score',y = 'GRE Score', hue='Research')
```

```python
X = df.iloc[:,0:2]
Y = df.iloc[:,-3]
Y = Y.astype('int')
```

```python
from sklearn.linear_model import Perceptron
p = Perceptron()
```

```python
p.fit(X, Y)
```

```python
p.coef_
```

```python
p.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X.values, Y.values, clf = p, legend = 2)
```
